```python
import numpy as np
import pandas as pd

csv = pd.read_csv('application_train.csv')
df = pd.DataFrame(csv)
print(df.head())

```

```python
##sklearn random split
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
train.head()
test.head()
```

```python
##from scratch random split
def random_split(dataframe, test_size):
    df_shuffled = dataframe.sample(frac = 1, random_state = 42).reset_index(drop=True)
    index_split = int(len(df)*(1 - test_size))
    df_train = df_shuffled.iloc[:index_split]
    df_test = df_shuffled.iloc[index_split:]
    return df_train, df_test
    
train, test = random_split(df, 0.2)
```
