This is under the assumption that SK_ID_CURR ids are time-based - that is, a loan with SK_ID_CURR of 100043 must have taken place after SK_ID_CURR 10002.

```python
import pandas as pd
import numpy as np
```

```python
# This can be substituted for any cleaned/feature engineered data frame
df = pd.read_csv('../rory_work/application_train.csv')
```

```python
def sorted_split(data, sort_col, train_size, ascending=True):
    # Sorts data along sort_col, then splits it,
    # placing train_size in the train set and 1-train_size in the test split
    # Returns both the train and test splits
    data.sort_values(sort_col, ascending=ascending)
    split = int(np.floor(data.shape[0] * train_size))
    train = data.iloc[:split, :]
    test = data.iloc[split:, :]
    
    return train, test
```

```python
train, test = sorted_split(df, "SK_ID_CURR", 0.8)
```
